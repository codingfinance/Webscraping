# Web scraping S&P tickers

The below code can be used to download the S&P 500 companies data.

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup

wiki_url = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
wiki_url = wiki_url.text

soup = BeautifulSoup(wiki_url, 'html.parser')

tbody = soup.tbody

tr = tbody.find_all('tr')

data = []

for t in tr:
    data.append(t.text.split('\n'))

raw_df = pd.DataFrame(data)

raw_df.columns = raw_df.iloc[0,:]

raw_df = raw_df.iloc[1:,:]
```
