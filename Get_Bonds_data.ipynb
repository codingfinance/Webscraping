```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
import re
```

```python
my_url = "https://markets.businessinsider.com/bonds/finder?p=1&borrower=&maturity=&yield=&bondtype=6,7,8,19&coupon=&currency=333&rating=&country="
```

```python
req = requests.get(my_url)
```

```python
soup = BeautifulSoup(req.text, 'lxml')
```

```python
table = soup.find("table")
```

```python
links = [i['href'] for i in table.find_all('a', href = re.compile('^/bonds'))]
links = [str('https://markets.businessinsider.com') + i for i in links]
name = [str(i.text).strip() for i in table.select('td a')]
currency = [str(i.text).strip() for i in table.select('.text-right:nth-child(2)')][1:]
coupon = [str(i.text).strip() for i in table.select('.text-right:nth-child(3)')][1:]
yld = [str(i.text).strip() for i in table.select('.text-right:nth-child(4)')][1:]
rtn = [str(i.text).strip() for i in table.select('.text-right:nth-child(5)')][1:]
maturity = [str(i.text).strip() for i in table.select('.text-right:nth-child(6)')][1:]
bid = [str(i.text).strip() for i in table.select('.text-right:nth-child(7)')][1:]
ask = [str(i.text).strip() for i in table.select('.text-right:nth-child(8)')][1:]
```

```python
df = pd.DataFrame({'name':name,
             'currency':currency,
             'coupon':coupon,
             'yld':yld,
             'rtn':rtn,
             'maturity':maturity,
             'bid':bid,
             'ask':ask, 'link':links})
```
