```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from selenium import webdriver

driver = webdriver.Firefox()
driver.get("https://www.preferredstockchannel.com/offers/?email=jacobjavin@me.com&first_name=Jacob&last_name=Javin&pid=&redir=https%3A%2F%2Fwww.preferredstockchannel.com%2Fslideshows%2F%3Fslideshow%3Dspx%26page%3D1%26confirm%3D1")


# "https://www.preferredstockchannel.com/offers/?email=jacobjavin@me.com&first_name=Jacob&last_name=Javin&pid=&redir=https%3A%2F%2Fwww.preferredstockchannel.com%2Fslideshows%2F%3Fslideshow%3Dspx%26page%3D1%26confirm%3D1%26confirm%3D1"
```

```python
import time

names = []

my_dict = []

for i in np.arange(1,8):
    
    my_url = "https://www.preferredstockchannel.com/slideshows/?slideshow=spx&page="+str(i)+"&confirm=1"
    
    driver.get(my_url)

    soup = BeautifulSoup(driver.page_source, "lxml")
       
    names.append(soup.select('h1')[1].text)
    
    data = [i.text for i in soup.select(".dstyb")]
    
    cols = [i.text for i in soup.select(".dsty")]
    
    _dict = dict(zip(cols,data))
    
    my_dict.append(_dict)
    
    sleep_time = np.random.randint(low = 1.02,
                 high = 3.25)
    
    time.sleep = sleep_time
    
    print("Done getting data for page", i)
```

```python
driver.close()
```

```python
my_df = pd.concat([pd.Series(d) for d in my_dict], axis=1, sort = False).fillna(0).T
```

```python
my_df['Name'] = names
```
